```python
from pyrovision.datasets import OpenFire
from torchvision import transforms as tf
from matplotlib import pyplot as plt
from PIL import Image
import torch
import os
```

```python
train_set = OpenFire(root="openfire", train=True, download=True)
print(f"Train Set Length: {len(train_set)}")
```

```python
# Shuffle train_set
shuffled_train_set = torch.utils.data.Subset(train_set, torch.randperm(len(train_set)))

# Visualize 12 images in a grid using matplotlib
fig, axs = plt.subplots(3, 4, figsize=(10, 10))
for i in range(12):
    img, target = shuffled_train_set[i]
    label = "Wild Fire" if target == 1 else "Normal"
    axs[i // 4, i % 4].imshow(img)
    axs[i // 4, i % 4].set_title(f"{label}")
    axs[i // 4, i % 4].axis("off")

plt.show()
```
