```python
import os
import random
from matplotlib import pyplot as plt

base_dir = "/home/umar/Documents/uwaterloo/4b/capstone_firefly/local_sandbox/FLAME"
image_base_dir = os.path.join(base_dir, "Images")
mask_base_dir = os.path.join(base_dir, "Masks")

image_list = [os.path.join(image_base_dir, i) for i in os.listdir(image_base_dir) if i.endswith(".jpg")]
mask_list = [os.path.join(mask_base_dir, i) for i in os.listdir(mask_base_dir) if i.endswith(".png")]

image_list.sort()
mask_list.sort()

# Zip the image and mask paths together
image_mask_list = list(zip(image_list, mask_list))

print("Number of image/mask pairs: ", len(image_mask_list))

# Print the paths of the first 5 images and masks
print("Image and mask paths: ", image_mask_list[:5])
```

```python
# Create a shuffled copy of the image/mask list
random.seed(42)
random.shuffle(image_mask_list)

# Visualize 6 pairs of images and masks next to each other using matplotlib
fig, ax = plt.subplots(3, 4, figsize=(20, 10))
for i in range(3):
    for j in range(4):
        ax[i, j].imshow(plt.imread(image_mask_list[i*4+j][0]))
        ax[i, j].imshow(plt.imread(image_mask_list[i*4+j][1]), alpha=0.5)
        ax[i, j].axis("off")
plt.show()
```
